````python
df_na = df.dropna()
df_na
````

````python
# Convert the categorical to numerical column, i.e. ocean_proximity
df_na.ocean_proximity
````

````python
df_na.ocean_proximity.value_counts()
````

````python
# TONOTE: `get_dummies` is used to convert categorical data into numerical data
dummies = pd.get_dummies(df_na.ocean_proximity, dtype=int)
dummies
````

````python
# Join the dummies with the df_na dataframe
df_na = df_na.join(dummies)
````

````python
df_na
````

````python
# Remove the ocean_proximity categorical column after joining its dummies to the dataframe
df_na = df_na.drop("ocean_proximity", axis=1)
````

````python
df_na
# TONOTE: Now the dataframe consists of only numerical columns, which is ready to implement any machine learning models. Machine learning only works with numerical columns
````

````python
# Check the data's correlations
df_na.corr()
````

````python
# The above correlation is confusing. Let's do it using seaborn for better visualization

sns.set_theme(rc={'figure.figsize': (15, 8)}) # `rc` stands for `reconfigure`
sns.heatmap(df_na.corr(), annot=True, cmap='YlGnBu')
````
